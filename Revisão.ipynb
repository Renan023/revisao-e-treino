```python
wine = datasets.load_wine()#guardar a informação do sklearn datasets na variavel 
x = wine.data#dados
y = wine.target#quantidade de amostras
```

```python
wine.target_names#nomes das amostras 
```

```python
wine.feature_names#quantidade dos dados da amostra 
```

```python
wine.data.shape#dados da quantidade e dados da amostra 
```

```python
vinho = pd.DataFrame(x,columns=wine.feature_names)#criação de um dataframe puxando a variável e  mostrando em colunas específicas 
vinho['target'] = pd.Categorical.from_codes(y,wine.target_names)#criação de uma coluna nova puxando a variável o nome dos seus dados 
```

```python
dfx = pd.DataFrame(x,columns=['alcohol',
 'malic_acid',
 'ash',
 'alcalinity_of_ash',
 'magnesium',
 'total_phenols',
 'flavanoids',
 'nonflavanoid_phenols',
 'proanthocyanins',
 'color_intensity',
 'hue',
 'od280/od315_of_diluted_wines',
 'proline'])#Cria cada coluna dando um resultado específico
dfy = pd.DataFrame(y, columns=['target'])
```

```python
wine = pd.concat([dfx,dfy],axis=1.0)#junção das variáveis 
```

```python
sns.pairplot(vinho,hue='target')#representação gráfica 
```

```python
x_train ,x_test , y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=100)
#Criação para o fatiamento para a variáveis para treino e teste com um valor determinado de amostras e aleatória 
```

```python
clf = svm.SVC(kernel='linear',C=1.0)#SuportVetorialMachine.Determina a acurácia, precisão 
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_train)#predizer dentro de uma variável 
```

```python
print(classification_report(y_train,y_pred))
```

```python
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head()
```

```python
df['Tempo de Entrega'] = (df['Data Envio']-df['Data Venda']).dt.days#cria uma coluna com a fórmula específica em calculando os dias 
```

```python
df.dtypes#mostra que tipo de dados está inserido na tabela 
```

```python
df['ID Loja'] = df['ID Loja'].astype('object')
df['ID Produto'] = df['ID Produto'].astype('object')
df['ID Cliente'] = df['ID Cliente'].astype('object')#pega as colunnas específicas e muda o tipo primitivo 
```

```python
df.dtypes
```

```python
pd.options.display.float_format = '{:20,.2f}'.format#Captura a tabela e arredonda os valores
```

```python
df.groupby([df['Data Venda'].dt.year,'Marca'])['Preço Unitário'].sum()#analisa uma determinada coluna e devolve a soma total 
```

```python
df.groupby([df['Data Venda'].dt.month,'Marca'])['Quantidade'].sum()
```

```python
df['Valor total'] = df['Preço Unitário']*df['Quantidade']
```

```python
df['Diferença'] = df['Preço Unitário'] - df['Custo Unitário']
```

```python
df['Lucro Total'] = df['Diferença']*df['Quantidade']
```
